```python
from google.colab import userdata
# Certifique-se de criar o segredo 'GROQ_API_KEY' no menu lateral (ícone de chave)
try:
    groq_api_key = userdata.get('GROQ_API_KEY')
    print('Chave recuperada com sucesso!')
except userdata.SecretNotFoundError:
    print('Erro: A chave GROQ_API_KEY não foi encontrada nos Segredos do Colab.')
```

```python
from llama_index.llms.groq import Groq
llm = Groq(model='llama-3.3-70b-versatile', api_key=groq_api_key)
response = llm.complete('Como ligar o temporizador da COEL TW21 24VCC?')
response
paragrafos = response.text.split("\n\n")
for paragrafo in paragrafos:
  print (paragrafo)
  print()
```

```python
from llama_index.llms.groq import Groq
llm = Groq(model='llama-3.3-70b-versatile', api_key=groq_api_key)
while True:
  response = llm.complete(input())
  response
  paragrafos = response.text.split("\n\n")
  for paragrafo in paragrafos:
    print(paragrafo)
    print()
```
