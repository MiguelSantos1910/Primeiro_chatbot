```python
from llama_index.llms.groq import Groq
from llama_index.core.base.llms.types import ChatMessage, MessageRole
```

```python
from google.colab import userdata
# Certifique-se de criar o segredo 'GROQ_API_KEY' no menu lateral (ícone de chave)
try:
    groq_api_key = userdata.get('GROQ_API_KEY')
    print('Chave recuperada com sucesso!')
except userdata.SecretNotFoundError:
    print('Erro: A chave GROQ_API_KEY não foi encontrada nos Segredos do Colab.')
```

```python
while True:
  user_prompt = input(" \n Digite sua pergunta: ")
  client = Groq(model="llama-3.3-70b-versatile", api_key=groq_api_key)
  completion = client.chat(
      messages=[
          ChatMessage(role=MessageRole.SYSTEM, content="Você é um assistente de agricultor"),
          ChatMessage(role=MessageRole.USER, content=user_prompt)
      ],
      temperature = 1,
      max_tokens = 1024,
      top_p = 1,
  )
  print("\n Resposta do assistente: \n ")
  print(completion.message.content)
```

```python
import gradio as gr

def generate_response(user_question):
    client = Groq(model="llama-3.3-70b-versatile", api_key=groq_api_key)
    completion = client.chat(
        messages=[
            ChatMessage(role=MessageRole.SYSTEM, content="Você é um assistente de agricultor"),
            ChatMessage(role=MessageRole.USER, content=user_question)
        ],
        temperature = 1,
        max_tokens = 1024,
        top_p = 1,
    )
    return completion.message.content

face = gr.Interface(
    fn=generate_response,
    inputs = gr.Textbox(lines = 3, placeholder="Digite sua pergunta sobre agricultura"),
    outputs = gr.Textbox(lines = 3, placeholder="Resposta do assistente"),
    title = "Assistente de Agricultor",
    description = "Digite sua pergunta e o assistente responderá",
    live = False
)

face.launch()
```
